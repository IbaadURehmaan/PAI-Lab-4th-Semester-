#### Unique Mathematical Backtracking Logic

```python
def is_queen_safe(current_row, test_col, past_placements):
    for r in range(current_row):
        c = past_placements[r]
        
        if c == test_col:
            return False
        
        row_diff = current_row - r
        col_diff = test_col - c
        
        if col_diff < 0:
            col_diff = -col_diff
            
        if row_diff == col_diff:
            return False
            
    return True

def find_all_arrangements(board_dim, current_r, placements, all_found):
    if current_r == board_dim:
        valid_layout = []
        for row_idx in range(board_dim):
            queen_c = placements[row_idx]
            row_string = ""
            
            for col_idx in range(board_dim):
                if col_idx == queen_c:
                    row_string += "Q "
                else:
                    row_string += ". "
                    
            valid_layout.append(row_string.strip())
            
        all_found.append(valid_layout)
        return

    for attempt_c in range(board_dim):
        if is_queen_safe(current_r, attempt_c, placements):
            placements[current_r] = attempt_c
            
            find_all_arrangements(board_dim, current_r + 1, placements, all_found)
            
            placements[current_r] = -1

def solve_n_queens_task(n_value):
    successful_boards = []
    initial_tracker = [-1] * n_value
    
    find_all_arrangements(n_value, 0, initial_tracker, successful_boards)
    
    return successful_boards
```

#### Execution and Output

```python
size_of_board = 4
final_answers = solve_n_queens_task(size_of_board)

print(f"Board Size: {size_of_board}x{size_of_board}")
print(f"Total valid arrangements found: {len(final_answers)}\n")

for idx, board in enumerate(final_answers):
    print(f"Arrangement {idx + 1}:")
    for line in board:
        print(line)
    print("")
```
