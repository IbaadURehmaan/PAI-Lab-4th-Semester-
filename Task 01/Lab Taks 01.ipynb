#### Setup & Data Loading

```python
import pandas as pd

from sklearn.tree import DecisionTreeRegressor


print("Loading the datasets...")


# reading the csv files
train_df = pd.read_csv('train.csv')

test_df = pd.read_csv('test.csv')


print("Data loaded successfully!")
```

#### Selecting Features

```python
my_features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']


# setting up X (features) and y (target/price)

X_train = train_df[my_features]

y_train = train_df['SalePrice']


# test data features

X_test = test_df[my_features]


print(f"Features we are using for prediction: {my_features}")
```

#### Training the AI Model

```python
house_model = DecisionTreeRegressor(random_state=42)

print("Training the model with our data. Please wait...")

house_model.fit(X_train, y_train)

print("Training done! The model is ready to make predictions.")
```

#### Predicting & Making the Submission File

```python
predictions = house_model.predict(X_test)

my_submission = {

    'Id': test_df['Id'],

    'SalePrice': predictions
}


final_output = pd.DataFrame(my_submission)
final_output.to_csv('submission.csv', index=False)

print("submission.csv is saved in the folder.")
print("Now it is ready to uplaod to kaggle competition page.")
```
