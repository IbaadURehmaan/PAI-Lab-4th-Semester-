#### Reading, Resizing, Cropping aur Displaying Images

```python
import cv2

my_img = cv2.imread('download.jpg')

resized_img = cv2.resize(my_img, (500, 500))

cropped_img = resized_img[50:400, 50:400]

cv2.imshow('Original Resized', resized_img)

cv2.imshow('Cropped Version', cropped_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
